```python
carousel_data = pd.DataFrame()
date = []
for i in tqdm(range (carousel.shape[0])):
    a = carousel['wtl'][i][2:-2]
    b = a.split(",")
    c = []
    for k in range(len(b)):
        temp = b[k].split(':')
        c.append(int(temp[1][:-1]))
    pd.DataFrame(c)
    carousel_data = carousel_data.append(c,ignore_index=True)
    s = carousel['created_on'][i][:10]
    for j in range(15):
        date.append(int(s.replace('-','')))   
carousel_data['date'] = pd.DataFrame(date)
```

```python
i1 = (carousel_data['date'].apply(lambda x: int(x)) < 20190625)
i2 = (carousel_data['date'].apply(lambda x: int(x)) > 20190501)
i3 =i1 & i2
x =carousel_data.loc[i3]
```

```python
x.sort_values([0 , 'date'], inplace = True)
x = x.reset_index(drop=True)
```

```python
crdata1 = x.drop_duplicates([0], keep = 'last')
crdata1 = crdata1.reset_index(drop=True)
```

```python
crdata = x.drop_duplicates([0], keep = 'first')
crdata = crdata.reset_index(drop=True)
```

```python
import tqdm
from tqdm import tqdm

rawparams = pd.DataFrame(index = [0],columns=['articleid','start','duration']).fillna(0)
for i in tqdm(range(crdata.shape[0])):
    artid = crdata[0][i]
    for j in range(crdata1.shape[0]):
        if crdata1[0][j] == artid:
            x1 = crdata1['date'][j] - crdata['date'][i]
    temp = pd.DataFrame(np.column_stack([artid,crdata['date'][i],x1]),columns=['articleid','start','duration'])
    rawparams = rawparams.append(temp,ignore_index=True)
```
