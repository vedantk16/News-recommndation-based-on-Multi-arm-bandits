```python
import numpy as np
import pandas as pd
import os
import json
import avro.schema
from avro.datafile import DataFileReader, DataFileWriter
from avro.io import DatumReader, DatumWriter
```

```python
reader = DataFileReader(open("./asiaville/20190315044423-divolte-tracking-ip-172-31-21-126-2.avro","rb",buffering = 0), DatumReader())
for user in reader:
    print(user)
reader.close()
```

```python
dir0= './asiaville'
a = os.listdir(dir0)
j=1

eventType = pd.DataFrame(index = [0],columns=['date','timestamp','user1', 'eventType']).fillna(0)
for k in a:
    if k[:4] == '2019':
        b = []
        reader = DataFileReader(open(os.path.join(dir0,k),"rb",buffering = 0), DatumReader())
        for user in reader:
            b.append(user)
        reader.close()
        b = pd.DataFrame(b)
        x = b['eventType'].values.tolist()
        y = [j for i in range(len(x))]
        z = b['timestamp'].values.tolist()
        k = [k[:8] for i in range(len(x))]
        data = pd.DataFrame(np.column_stack([k,z,y, x]), columns=['date','timestamp','user1', 'eventType'])
        eventType = eventType.append(data,ignore_index=True)
        j=j+1
```

```python
eventType.shape
```

```python
i = (eventType['eventType'] != 'pageView') 
i1 = (eventType['eventType'] != 'bannerClick')
i2 =  (eventType['eventType'] != 0)
i4 = (eventType['date'].apply(lambda x: int(x)) > 20190312)
i3 = i & i1 & i2 & i4
x =eventType.loc[i3] 
```

```python
data = pd.DataFrame(index = [0], columns=['date','timestamp', 'user1', 'userid', 'articleid', 'carousel_index']).fillna(0)
for i in range(x.shape[0]):
    k = json.loads(x['eventType'].iloc[i])
    if ('articleid' in k.keys()) & ('userid' in k.keys()):
        x1 = x['date'].iloc[i]
        x2 = x['timestamp'].iloc[i]
        x3 = x['user1'].iloc[i]
        if k['userid'] == '':
            x4 = -1
        else:
            x4 = k['userid']
        x5 = k['articleid']
        x6 = k['carousel_index']
        d = pd.DataFrame(np.column_stack([x1,x2,x3,x4,x5,x6]), columns=['date','timestamp', 'user1', 'userid', 'articleid', 'carousel_index'])
        data = data.append(d,ignore_index=True)
```
